```python
#Second method
#This a different method where I am trying to define everything and call the functions
#I keep getting a syntax error but I wanted you to look at it to see if It makes sense
# create age dictionary
age_dict ={2:0.171916334485, 3:0.112524355352, 4:0.010417899096, 5:-0.086398020160, 
           6:-0.330472881810, 7: -0.238477625234, 8:-0.381620279917, 9:-0.153686593163,
           10: -0.186548951503, 11:-0.083139710082, 12:-0.056656757634, 13:-0.284882801544,
            14:-0.396864965893, 15:-0.568987961763, 16:-0.727763107647, 17:-0.850963121903,
            18:-0.978545665276, 19:-1.302052025880, 20:0.488821830927}

#this equation is included in all the alternatives on which to add on depending on the alternatives

def base_eqn():
    base =  veh910.loc['i025_age', 'Coefficient'] * NHTS['INC025']+veh910.loc['i2550_age', 'Coefficient'] * NHTS['INC2550']+ veh910.loc['i100_age', 'Coefficient'] * NHTS['INC100150']+ veh910.loc['i150p_age', 'Coefficient']  * NHTS['INC150P']+veh910.loc['imiss_age', 'Coefficient']  * NHTS['INC_MISS']+veh910.loc['oneveh_age', 'Coefficient'] * NHTS['ONEVEH']+ veh910.loc['vhgtdr_age', 'Coefficient'] * NHTS['VHGTDR']+ veh910.loc['nchld_age', 'Coefficient']  * NHTS['NUM_CHILD3'] + veh910.loc['dstwkt_age', 'Coefficient'] * NHTS['DSTWKT']+ veh910.loc['sfo_age', 'Coefficient'] * NHTS['SFO']+veh910.loc['san_age', 'Coefficient'] * NHTS['SAN']+veh910.loc['atl_age', 'Coefficient'] * NHTS['ATL']+ veh910.loc['sea_age', 'Coefficient'] * NHTS['SEA']+ veh910.loc['det_age', 'Coefficient'] * NHTS['DET']+ veh910.loc['msp_age', 'Coefficient'] * NHTS['MSP']+ veh910.loc['dca_age', 'Coefficient'] * NHTS['DCA']+ veh910.loc['oreg_age', 'Coefficient'] * NHTS['OREG']+ veh910.loc['ohio_age', 'Coefficient'] * NHTS['OHIO']+ veh910.loc['rural_age', 'Coefficient'] * NHTS['RURAL']+ veh910.loc['cmpg', 'Coefficient'] * 24.0 + veh910.loc['cprice0', 'Coefficient']* NHTS['INC025']* 32927+veh910.loc['cprice25', 'Coefficient'] * NHTS['INC2550'] * 32927 +veh910.loc['cprice50', 'Coefficient'] * NHTS['INC50100'] * 32927 + veh910.loc['cprice100', 'Coefficient'] * NHTS['INC100150'] * 32927 +veh910.loc['cprice150', 'Coefficient'] * NHTS['INC150P'] * 32927+ veh910.loc['cpricem', 'Coefficient'] * NHTS['INC_MISS'] * 32927
    return base


def get_vehcoeff(_vehtype,_houseid):
    if _vehtype=="van":
        stype ="van"
    elif _vehtype =="suv":
        stype="suv"
    elif _vehtype == "pickup":
        stype = "pu"  
    elif _vehtype == "motorcycle":
        stype = "mc"
    
    
    coef = veh910.loc[f"{stype}"] + veh910.loc[f"vhgtdr_{stype}"]*NHTS.loc[NHTS["HOUSEID"]==_houseid,"VHGTDR"] + 
            veh910.loc[f"nchld_{stype}"]*NHTS.loc[NHTS["HOUSEID"]==_houseid,'NUM_CHLD3'] + 
            veh910.loc[f"den3_{stype}"]*NHTS.loc[NHTS["HOUSEID"]==_houseid,'DEN3'] +
            veh910.loc[f"den4_{stype}"]*NHTS.loc[NHTS["HOUSEID"]==_houseid, DEN4 + f"den5_{stype}"*DEN5 + f"den6_{stype}"*DEN6 + 
            f"den7_{stype}"*DEN7 +f"den8_{stype}"*DEN8 + f"sfo_{stype}"*SFO + f"san_{stype}"*SAN + 
            f"atl_{stype}"*ATL + f"sea_{stype}"*SEA + f"det_{stype}"*DET + f"msp_{stype}"*MSP + 
            f"dca_{stype}"*DCA +f"oreg_{stype}"*OREG + f"ohio_{stype}"*OHIO 
            
    return coef
    
# for i in EPA:
#     to_array = list(EPA["CHOICE4"])   


#extra coefficients for ev's
def get_coeff_PEV_BEV_hybrid(_fueltypename):    
    if _fueltypename=="PEV":
        stype ="ev"
    elif _fueltypename =="BEV":
        stype="ev"
    elif _fueltypename == "hybrid":
        stype = "hyb"   
    
    coef = stype + f"vhgtdr_{stype}"*VHGTDR + f"nchld_{stype}"*NUM_CHLD3 + f"den3_{stype}"*DEN3 + f"den4_{stype}"*DEN4 + f"den5_{stype}"*DEN5 + f"den6_{stype}"*DEN6 + f"den7_{stype}"*DEN7 +f"den8_{stype}"*DEN8 + f"sfo_{stype}"*SFO + f"san_{stype}"*SAN + f"atl_{stype}"*ATL + f"sea_{stype}"*SEA + f"det_{stype}"*DET + f"msp_{stype}"*MSP + f"dca_{stype}"*DCA +f"oreg_{stype}"*OREG + f"ohio_{stype}"*OHIO + f"hyb_{stype}"* OWN_HYB
        
    return coef

def get_fuelcoeff(_fueltype):
    fuel_type=_fueltype
    fuel_coeff = 0
    if fuel_type=="gas":
        #no extra coefficient
        fuel_coeff = 0
    elif fuel_type=="diesel":
        fuel_coeff =veh910.loc['dsl', 'Coefficient']
    elif fuel_type=="hybrid":
        #add hybrid coefficient
        fuel_coeff = get_coeff_PEV_BEV_hybrid(fuel_type)         
    elif fuel_type=="PEV":
        #add pev coefficient
        fuel_coeff = get_coeff_PEV_BEV_hybrid(fuel_type)
    elif fuel_type=="BEV":
        #add bev coefficient
        fuel_coeff = get_coeff_PEV_BEV_hybrid(fuel_type)        
    return fuel_coeff
        
def get_fueltype(to_array):
    fuel_type= ""
    if to_array[1]=="1":
        fuel_type="gas"        
    elif to_array[1]=="2":
        fuel_type="diesel"        
    elif to_array[1]=="3":
        fuel_type="hybrid"                 
    elif to_array[1]=="4":
        fuel_type="PEV"        
    elif to_array[1]=="5":
        fuel_type="BEV"        
        
    
    return get_fuelcoeff(fuel_type)






#total utility equations
def utilty_eqn(_vehtype, _fueltype, _age):
    
    ## include the base equation
    base_coeff = base_eqn()
    ## include the vehtype equation
    veh_coeff = get_vehcoeff(_vehtype)
    ## get the fuel special equations
    fuel_coeff = get_fueltype(_fueltype)
    # get age coeff
    age_coeff = age_dict[age]
    
    coeff = base_coeff + veh_coeff + fuel_coeff + age_coeff
    return coeff


#here a for loop is made to assign each number of the choice to their alternatives
#for example car is 1, car-diesel is 12, car-diesel-age6 is 1206
for ind, row in EPA.iterrows():
    to_array = list(row["CHOICE4"])
    if to_array[0] == "1":
        ## this is car
        veh_type = "car"
        #no extra coefficient
        veh_coeff = 0
        #find fuel type
        fuel_type = get_fueltype(to_array)
        #calculate age and get age coefficient
        age = int(to_array[2] + to_array[3])
        if age<=1:
            age_coef=0
        else:
            age_coef = age_dict[age]        
                        
    elif to_array[0]=="2":
        ## this is VAN
        veh_type = "van"
        #get van coefficient equations
        veh_coeff = get_vehcoeff(_vehtype,row["HOUSEID"])
        #find fuel type
        fuel_coeff = get_fueltype(to_array)
        #calculate age and get age coefficient
        age = int(to_array[2] + to_array[3])
        if age<=1:
            age_coef=0
        else:
            age_coef = age_dict[age]
                        
    elif to_array[0]=="3":
        ## this is SUV
        veh_type = "suv" 
        #get suv coefficient equations
        veh_coeff = get_vehcoeff(_vehtype)
        #find fuel type
        fuel_type = get_fueltype(to_array)
        #calculate age and get age coefficient
        age = int(to_array[2] + to_array[3])
        if age<=1:
            age_coef=0
        else:
            age_coef = age_dict[age]            
            
    elif to_array[0]=="4":
        ## this is Pickup
        veh_type = "pickup"
        #get pickup coefficient equations
        veh_coeff = get_vehcoeff(_vehtype)
        #find fuel type
        fuel_type = get_fueltype(to_array)
        #calculate age and get age coefficient
        age = int(to_array[2] + to_array[3])
        if age<=1:
            age_coef=0
        else:
            age_coef = age_dict[age]
                        
    elif to_array[0]=="5":
        #this is motorcycle
        veh_type = "motorcycle"
        #get motorcycle coefficient equations
        veh_coeff = get_vehcoeff(_vehtype)
        #find fuel type
        fuel_type = get_fueltype(to_array)
        #calculate age and get age coefficient
        age = int(to_array[2] + to_array[3])
        if age<=1:
            age_coef=0
        else:
            age_coef = age_dict[age]
    
    # finally do this
    coeff = utilty_eqn(veh_type,fuel_type,age)
    row["coeff"] = coeff
        




# coeff = utilty_eqn(veh_type,fuel_type,age)
    
    
    
    
```
